```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data/dataset_1st/training.csv')
df.dropna()
df
```

```python
X1 = df['conam'].values
X2 = df['ecfg'].values
X3 = df['insfg'].values
X4 = df['iterm'].values
X5 = df['bnsfg'].values
X6 = df['flam1'].values
X7 = df['stocn'].values #
X8 = df['scity'].values #
X9 = df['stscd'].values #
X10 = df['ovrlt'].values
X11 = df['flbmk'].values
X12 = df['hcefg'].values #
X13 = df['csmcu'].values #
X14 = df['csmam'].values
X15 = df['flg_3dsmk'].values
Y = df['label'].values

X = np.array(list(zip(X1,X2,X3,X4,X5,X6,X10,X11,X14,X15)))
#X = X.reshape(-1,1)


model = linear_model.LogisticRegression()
model.fit(X, Y)

prediction = model.predict(X)

accuracy_train = model.score(X, Y)


coefficients = list(model.coef_[0])
intercept = model.intercept_[0]

print("coefficient: ", coefficients)
print("intercept: ", intercept)
```

```python
def predict_equation(x1,x2,x3,x4,x5,x6,x10,x11,x14,x15):
    value = (x1 * coefficients[0]) + (x2 * coefficients[1]) + (x3 * coefficients[2]) + (x4 * coefficients[3]) + (x5 * coefficients[4]) + (x6 * coefficients[5]) + (x10 * coefficients[6]) + (x11 * coefficients[7]) + (x14 * coefficients[8]) + (x15 * coefficients[9]) + intercept
    
    if value < 0.5:
        return 0
    else:
        return 1
    
```

```python
df2 = pd.read_csv('data/dataset_1st/public_processed.csv')
df2
```

```python
T1 = df2['locdt']
T2 = df2['loctm']
T3 = df2['conam']
T4 = df2['iterm']
T5 = df2['flam1']
T6 = df2['csmam']
T7 = df2['stocn'] #
T8 = df2['scity'] #
T9 = df2['stscd'] #
T10 = df2['ovrlt']
T11 = df2['flbmk']
T12 = df2['hcefg'] #
T13 = df2['csmcu'] #
T14 = df2['csmam']
T15 = df2['flg_3dsmk']
predict_result = []


for i in range(0,len(df2)):
    predict_result.append(predict_equation(T1[i],T2[i],T3[i],T4[i],T5[i],T6[i],T10[i],T11[i],T14[i],T15[i]))

print(predict_result)



```
