```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

# Data Preparation

```python
df = pd.read_csv('data/dataset_1st/training.csv')
```

```python
df.isnull()
```

```python
df.isna().any()
```

```python
dropCol = ['chid', 'cano', 'mchno', 'acqic', 'stscd']
df = df.drop(columns = dropCol)
df = df.dropna()
df.isna().any()
```

```python
df_label_0 = df[df['label'] == 0]
df_label_0
```

```python
df_label_1 = df[df['label'] == 1]
df_label_1
```

```python
label_0_rows = len(df_label_0)
label_1_rows = len(df_label_1)
min_rows = min(label_0_rows, label_1_rows)

df_label_0_sample = df_label_0.sample(n = min_rows, random_state = 42)
df_label_1_sample = df_label_1.sample(n = min_rows, random_state = 42)

balanced_df = pd.concat([df_label_0_sample, df_label_1_sample])
balanced_df
```

```python
balanced_df.keys()
```

```python
X1 = balanced_df['locdt'].values      #授權日期
X2 = balanced_df['loctm'].values      #授權時間
X3 = balanced_df['contp'].values      #交易類別
X4 = balanced_df['etymd'].values      #交易型態
X5 = balanced_df['mcc'].values        #mcc_code
X6 = balanced_df['conam'].values      #交易金額-台幣
X7 = balanced_df['ecfg'].values       #網路交易註記
X8 = balanced_df['insfg'].values      #是否分期交易
X9 = balanced_df['iterm'].values      #分期期數
X10 = balanced_df['bnsfg'].values     #是否紅利交易
X11 = balanced_df['flam1'].values     #實付金額
X12 = balanced_df['stocn'].values     #消費地國別
X13 = balanced_df['scity'].values     #消費城市  
X14 = balanced_df['ovrlt'].values     #超額註記碼
X15 = balanced_df['flbmk'].values     #Fallback註記
X16 = balanced_df['hcefg'].values     #支付型態
X17 = balanced_df['csmcu'].values     #消費地幣別
X18 = balanced_df['csmam'].values     #消費地金額
X19 = balanced_df['flg_3dsmk'].values #3D交易註記
Y = balanced_df['label'].values       #盜刷與否

X = np.array(list(zip(X1,X2,X3,X4,X5,X6,X7,X8,X9,X10,X11,X12,X13,X14,X15,X16,X17,X18,X19)))
#X = X.reshape(-1,1)


model = linear_model.LogisticRegression(solver='lbfgs', max_iter = 100000)
model.fit(X, Y)

prediction = model.predict(X)

accuracy_train = model.score(X, Y)


coefficients = list(model.coef_[0])
intercept = model.intercept_[0]

print(("accuracy: %.3f " % (accuracy_train*100)) + "%")
print("coefficient: ", coefficients)
print("intercept: ", intercept)
```

```python
def predict_equation(x1,x2,x3,x4,x5,x6,x7,x8,x9,x10,x11,x12,x13,x14,x15,x16,x17,x18,x19):

    value = (x1 * coefficients[0]) + (x2 * coefficients[1]) + (x3 * coefficients[2]) + (x4 * coefficients[3]) + (x5 * coefficients[4]) + (x6 * coefficients[5]) + (x7 * coefficients[6]) + (x8 * coefficients[7]) + (x9 * coefficients[8]) + (x10 * coefficients[9]) + (x11 * coefficients[10]) + (x12 * coefficients[11]) + (x13 * coefficients[12]) + (x14 * coefficients[13]) + (x15 * coefficients[14]) + (x16 * coefficients[15]) + (x17 * coefficients[16]) + (x18 * coefficients[17]) + (x19 * coefficients[18]) + intercept
    
    if value < 0.5:
        return 0
    else:
        return 1
    
```

```python
T1 = df2['locdt']
T2 = df2['loctm']
T3 = df2['contp']
T4 = df2['etymd']
T5 = df2['mcc']
T6 = df2['conam']
T7 = df2['ecfg']
T8 = df2['insfg']
T9 = df2['iterm']
T10 = df2['bnsfg']
T11 = df2['flam1']
T12 = df2['stocn']
T13 = df2['scity'] 
T14 = df2['ovrlt']
T15 = df2['flbmk']
T16 = df2['hcefg']
T17 = df2['csmcu'] 
T18 = df2['csmam']
T19 = df2['flg_3dsmk']
predict_result = []


for i in range(0,len(df2)):
    predict_result.append(predict_equation(T1[i],T2[i],T3[i],T4[i],T5[i],T6[i],T7[i],T8[i],T9[i],T10[i],T11[i],T12[i],T13[i],T14[i],T15[i],T16[i],T17[i],T18[i],T19[i]))

print(predict_result[:1000])

print("total frauds: %d" % sum(predict_result))
print("percentage of faruds: %.2f" % ((100 * sum(predict_result))/len(predict_result)) + "%")



```
