```python
dropCol = ['chid','cano', 'locdt', 'loctm', 'scity', 'flbmk','mchno','acqic']
train = train.drop(columns = dropCol)
pp = pp.drop(columns = dropCol)
train.dropna(subset = ['stocn'], inplace = True)
pp.dropna(subset = ['stocn'], inplace = True)
train.dropna(subset = ['mcc'], inplace = True)
pp.dropna(subset = ['mcc'], inplace = True)
train['stscd'].fillna(-1, inplace = True)
pp['stscd'].fillna(-1, inplace = True)
train.dropna(subset = ['stscd'], inplace = True)
pp.dropna(subset = ['stscd'], inplace = True)
train['stscd'].fillna(-1, inplace = True)
pp['stscd'].fillna(-1, inplace = True)
train = train.dropna()
pp = pp.dropna()
train_mode = train['hcefg'].mode().iloc[0]
pp_mode = pp['hcefg'].mode().iloc[0]
train['hcefg'].fillna(train_mode, inplace = True)
pp['hcefg'].fillna(pp_mode, inplace = True)
most_common_currency_by_country = pp.groupby('stocn')['csmcu'].apply(lambda x: x.mode().iloc[0]).to_dict()
pp['csmcu'] = pp['stocn'].map(most_common_currency_by_country).fillna(pp['csmcu'])
T_most_common_currency_by_country = train.groupby('stocn')['csmcu'].apply(lambda x: x.mode().iloc[0]).to_dict()
train['csmcu'] = train['stocn'].map(T_most_common_currency_by_country).fillna(train['csmcu'])
imputec = ['etymd']
ipt_train = train[imputec].values.reshape(-1, 1)
ipt_pp = pp[imputec].values.reshape(-1, 1)
knn_imputer = KNNImputer(n_neighbors = 4)
civ = knn_imputer.fit_transform(ipt_pp)
pp['etymd'] = civ
train.dropna(subset = ['etymd'], inplace = True)
txkey = pp['txkey']
train = train.drop(columns = 'txkey')
pp = pp.drop(columns = 'txkey')
train_label_0 = train[train['label'] == 0]
train_label_1 = train[train['label'] == 1]
label_0_rows = len(train_label_0)
label_1_rows = len(train_label_1)
min_rows = min(len(pp), len(train)) - len(train_label_1)
train_label_0_sample = train_label_0.sample(n = min_rows, random_state = 42)
train_label_1_sample = train_label_1.sample(n = len(train_label_1), random_state = 42)
balanced_train = pd.concat([train_label_0_sample, train_label_1_sample])
balanced_train = balanced_train.sample(frac = 1, random_state = 42).reset_index(drop = True)
#X = train
#Y = train["label"]
#X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
#scaler = StandardScaler()
#X_train_scaled = scaler.fit_transform(X_train)
#smote = SMOTE(random_state = 42)
#X_resampled, Y_resampled = smote.fit_resample(X_train, Y_train)
#X_resampled_scaled = scaler.transform(X_resampled)
#clf = KNeighborsClassifier(n_neighbors = 5)
#clf.fit(X_resampled_scaled, Y_resampled)
#X_test_scaled = scaler.transform(X_test)
#Y_pred = clf.predict(X_test_scaled)

#X = balanced_train.drop("label", axis = 1)  # Features
#Y = balanced_train["label"]  # Target variable
#X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
#scaler = StandardScaler()
#X_train_scaled = scaler.fit_transform(X_train)
#smote = SMOTE(random_state = 42)
#X_resampled, Y_resampled = smote.fit_resample(X_train_scaled, Y_train)
#clf = KNeighborsClassifier(n_neighbors = 5)
#clf.fit(X_resampled, Y_resampled)
#X_test_scaled = scaler.transform(X_test)
#Y_pred = clf.predict(X_test_scaled)
#X_processed_scaled = scaler.transform(pp)
#Y_pred_processed = clf.predict(X_processed_scaled)
#pp = pd.concat([txkey, pp], axis = 1)
#pp["pred"] = y_pred_processed
balanced_train
```

```python
X1 = balanced_train['contp'].values      #交易類別
X2 = balanced_train['etymd'].values      #交易型態
X3 = balanced_train['mcc'].values        #mcc_code
X4 = balanced_train['conam'].values      #交易金額-台幣
X5 = balanced_train['ecfg'].values       #網路交易註記
X6 = balanced_train['insfg'].values      #是否分期交易
X7 = balanced_train['iterm'].values      #分期期數
X8 = balanced_train['bnsfg'].values     #是否紅利交易
X9 = balanced_train['flam1'].values     #實付金額
X10 = balanced_train['stocn'].values     #消費地國別
X11 = balanced_train['stscd'].values     #狀態碼
X12 = balanced_train['ovrlt'].values     #超額註記碼
X13 = balanced_train['hcefg'].values     #支付型態
X14 = balanced_train['csmcu'].values     #消費地幣別
X15 = balanced_train['csmam'].values     #消費地金額
X16 = balanced_train['flg_3dsmk'].values #3D交易註記
Y = balanced_train['label'].values       #盜刷與否

X = np.array(list(zip(X1,X2,X3,X4,X5,X6,X7,X8,X9,X10,X11,X12,X13,X14,X15,X16)))
#X = X.reshape(-1,1)


model = linear_model.LogisticRegression(solver='lbfgs', max_iter = 100000)
model.fit(X, Y)


prediction = model.predict(X)


accuracy_train = model.score(X, Y)



coefficients = list(model.coef_[0])
intercept = model.intercept_[0]

print(("accuracy: %.3f " % (accuracy_train * 100)) + "%")
print("coefficient: ", coefficients)
print("intercept: ", intercept)

```

```python
def predict_equation(x1,x2,x3,x4,x5,x6,x7,x8,x9,x10,x11,x12,x13,x14,x15,x16):

    value = (x1 * coefficients[0]) + (x2 * coefficients[1]) + (x3 * coefficients[2]) + (x4 * coefficients[3]) + (x5 * coefficients[4]) + (x6 * coefficients[5]) + (x7 * coefficients[6]) + (x8 * coefficients[7]) + (x9 * coefficients[8]) + (x10 * coefficients[9]) + (x11 * coefficients[10]) + (x12 * coefficients[11]) + (x13 * coefficients[12]) + (x14 * coefficients[13]) + (x15 * coefficients[14]) + intercept
    
    if value < 0.5:
        return 0
    else:
        return 1
  
    
```
